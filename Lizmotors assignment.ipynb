```python
import requests
from bs4 import BeautifulSoup

url = 'https://investors.canoo.com/'

# Make a request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Now, you can use BeautifulSoup methods to navigate and extract information from the HTML structure.
    # For example, to find the title of the webpage:
    title = soup.title
    print("Title of the webpage:", title.text)

    # You can continue navigating the HTML structure and extract the information you need.

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
